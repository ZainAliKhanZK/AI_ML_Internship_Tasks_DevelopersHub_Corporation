Build a basic chatbot that provides supportive and empathetic responses for stress, anxiety,

## Problem Statement

Many individuals experiencing stress, anxiety, or emotional challenges may need immediate emotional support but do not always have access to professional help. 

This project aims to develop a chatbot that provides empathetic and supportive responses by learning from real human conversations.

## Goal

To fine-tune a small language model using the EmpatheticDialogues dataset and create a chatbot capable of generating emotionally supportive responses for mental health and emotional wellness conversations. 

The chatbot should maintain a compassionate, respectful, and safe tone while interacting with users.

-------------

```python
from app import generate_response
```

```python
print("----------Empathetic Chatbot----------\n— type 'quit' to exit.\n")

while True:
    user_input = input("You: ").strip()
    if user_input.lower() == "quit":
        break
    response = generate_response(user_input)
    print("Bot:", response, "\n")
```
