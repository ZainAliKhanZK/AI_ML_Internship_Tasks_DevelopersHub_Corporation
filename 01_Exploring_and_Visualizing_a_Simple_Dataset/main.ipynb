## Creating Scatter Plots

```python
s_length = df['SepalLengthCm']
s_width = df['SepalWidthCm']

p_length = df['PetalLengthCm']
p_width = df['PetalWidthCm']
ss = s_width * 6
```

```python
plt.scatter(s_length, s_width, s=ss, c=s_length, label='Sepal', cmap="viridis" ,edgecolor='black')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.colorbar(label='Sepal Length (cm)')
plt.grid(True, alpha=0.5)
plt.legend()
plt.show()
```

```python
ps = p_width * 10
```

```python
plt.scatter(p_length, p_width, s=ps, c=p_length, label='Petal', cmap="viridis" ,edgecolor='black')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.colorbar(label='Petal Length (cm)')
plt.grid(True, alpha=0.5)
plt.legend()
plt.show()
```

### Creating Histograms

```python
sepal_length = df['SepalLengthCm']

plt.hist(sepal_length, bins=10, edgecolor="black")
plt.grid(True, alpha=0.5)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
sepal_width = df['SepalWidthCm']

plt.hist(sepal_width, bins=10, edgecolor="black")
plt.grid(True, alpha=0.5)
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Width')
plt.show()
```

```python
petal_length = df['PetalLengthCm']

plt.hist(petal_length, bins=10, edgecolor="black")
plt.grid(True, alpha=0.5)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Length')
plt.show()
```

```python
petal_width = df['PetalWidthCm']

plt.hist(petal_width, bins=10, edgecolor="black")
plt.grid(True, alpha=0.5)
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Width')
plt.show()
```

```python
df.hist(figsize=(10,8), bins=15)

plt.suptitle("Distribution of Iris Features")
plt.show()
```

### - Creating a Boxplot

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
sns.boxplot(x = sepal_length , y = sepal_width , data = df)
plt.title("Box Plot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x = sepal_length , y = sepal_width , data = df)
plt.title("bar plot of sepal length vs sepal width".title())
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.lineplot(x = sepal_length , y = sepal_width , data = df)
plt.title("Line Plot of Sepal Length vs Sepal Width")
plt.show()
```

## Results and Final Insights

- Visualizations revealed clear differences among the three iris species based on sepal and petal measurements.
- Petal length and petal width were the most effective features for distinguishing between species.
- Strong relationships were observed between certain features, particularly petal length and petal width.
- The analysis demonstrated how data visualization helps uncover patterns, trends, and feature importance.
- Overall, the Iris dataset provides a simple yet effective example for learning exploratory data analysis (EDA) and understanding data-driven insights.
