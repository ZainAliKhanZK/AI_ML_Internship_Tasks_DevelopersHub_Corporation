## Problem Statement

People often seek quick answers to basic health-related questions, but inaccurate or unsafe information can lead to misunderstandings. This project aims to develop a health query chatbot that provides clear, informative, and safety-conscious responses using a Large Language Model (LLM).

## Goal

To build a chatbot that can answer general health-related questions using an LLM, apply prompt engineering techniques to generate friendly and understandable responses, and implement safety measures to prevent harmful or misleading medical advice.

------------

## Importing Libraries

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

## Getting the API key 

```python
load_dotenv()

api_key = os.getenv("HF")
```

## Creating a client of OpenAI

```python
client = OpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key=api_key,
)
```

## Using OpenAI's Chat Completion Model & 

## Getting the Response from Model

```python
print("-------- 🏥 Health Chatbot --------")
print("Type 'exit' to quit.\n")

while True:
    user_query = input("You: ")

    if user_query.lower() in ["exit", "quit", "bye"]:
        print("Bot: Take care! Goodbye. 👋")
        break

    completion = client.chat.completions.create(
        model="moonshotai/Kimi-K2-Instruct-0905",
        messages=[
            {
                "role": "system",
                "content": (
                    "You are a helpful and friendly medical assistant. "
                    "Do not provide harmful medical advice. "
                    "Encourage users to consult healthcare professionals for serious concerns. "
                    "Keep responses clear, concise, and well-structured."
                )
            },
            {
                "role": "user",
                "content": user_query
            }
        ]
    )

    response = completion.choices[0].message.content
    print(f"\nBot: {response}\n")
```

## Results and Final Insights

- A health query chatbot was successfully developed using an LLM and API integration.
- Prompt engineering techniques improved the clarity, helpfulness, and consistency of responses.
- Safety filters were implemented to avoid providing dangerous medical advice and to encourage users to consult healthcare professionals when necessary.
- The chatbot effectively answered common health-related questions while maintaining a supportive and informative tone.
- The project demonstrated the importance of prompt design, responsible AI usage, and safety mechanisms when developing healthcare-related conversational systems.
